# Конструктивы

Бывает такой тип задач, в которых надо **построить** какой-то объект, удовлетворяющий некоторым свойствам. Так называемые задачи-конструктивы.

Чаще всего они решаются с использованием идей **жадного алгоритма**. 

В этой теме не будет никаких конкретных алгоритмов, только разбор пары задач и несколько общих идей, здесь самое главное - порешать задачи из контеста, их много и они интересные.

Давайте разберем пару примеров таких задач.

## Пример
Надо построить такую последовательность клетчатых столбцов разных высот от 1 до N, чтобы, если налить сверху воды, то осталось бы ровно X квадратиков воды. А именно вода остается в квадратике в том случае, если и справа, и слева есть столбец на высоте этого квадратика.

Вот так, например, распределится вода, если поставить столбцы 1, 2, 5, 4, 10, 3, 6, 8, 9, 7:

|1|2|5|4|10|3|6|8|9|7|
|-|-|-|-|-|-|-|-|-|-|
| | | | |■| | | | | |
| | | | |■|□|□|□|■| |
| | | | |■|□|□|■|■| |
| | | | |■|□|□|■|■|■|
| | | | |■|□|■|■|■|■|
| | |■|□|■|□|■|■|■|■|
| | |■|■|■|□|■|■|■|■|
| | |■|■|■|■|■|■|■|■|
| |■|■|■|■|■|■|■|■|■|
|■|■|■|■|■|■|■|■|■|■|


### Задание
Чтобы решить задачу, сначала надо попробовать порешать ее на конкретных маленьких числах. Пусть у вас есть столбцы высоты 1, 2, 3, 4, 5, попробуйте поставить их в таком порядке, чтобы  осталось ровно 4 квадратика воды.

### Задание
А теперь попробуйте решить задачу для больших чисел: столбцы высоты от 1 до 10, и нужно построить так, чтобы было 24 квадратика воды.

Вообще полезно задать себе такой вопрос: а сколько минимально и максимально можно сделать квадратиков воды? Минимально можно ноль - если поставить все столбики по возрастанию, то вся вода точно стечет. А максимально?

Максимально получится, если два самых высоких столбца поставить как первый и последний, причем неважно, в каком порядке будут все между ними. В этом случается уровень воды будет равен $n-1$, и суммарно квадратиков воды будет $1 + 2 + \ldots + (n-2) = \frac{(n-2)(n-1)}{2}$. Больше воды не получится.

Логично предположить, что для всех X от $0$ до $\frac{(n-2)(n-1)}{2}$ действительно можно построить такую последовательность столбцов. Давайте попробуем это сделать **жадно** - почти всегда в конструктивах помогает жадность.

Давайте первым столбцом опять будет самый большой (n). И теперь будем пытаться сделать после него как можно более большую "яму", после которой поставим второй большой столбец (n - 1). Причем давайте постараемся сделать так, чтобы размер "ямы" был ровно X. Возьмем как второй столбец 1, после него 2 и так далее - при этом размер ямы увеличивается на (n - 2), потом на (n - 3) и так далее. В какой-то момент при добавлении нового столбца t яма стала больше, чем X. Тогда не нужно увеличивать так сильно. Давайте поставим столбец побольше, чтобы яма увеличась поменьше - ровно настолько, чтобы суммарно стала занимать X. После этого осталось поставить столбец (n-1) и справа от него поставить оставшиеся столбцы так, чтобы вся вода стекла точно - например, в убывающем порядке.

Заметьте, что столбец больше мы взять всегда можем - так как ставим столбцы начиная с 1 по возрастанию (не считая два самых больших). Вот какой столбец может получиться для столбцов от 1 до 10 и 24 квадратиков воды:

|**10**|1|2|3|6|**9**|8|7|5|4|
|-|-|-|-|-|-|-|-|-|-|
|■| | | | | | | | | |
|■|□|□|□|□|■| | | | |
|■|□|□|□|□|■|■| | | |
|■|□|□|□|□|■|■|■| | |
|■|□|□|□|■|■|■|■| | |
|■|□|□|□|■|■|■|■|■| |
|■|□|□|□|■|■|■|■|■|■|
|■|□|□|■|■|■|■|■|■|■|
|■|□|■|■|■|■|■|■|■|■|
